```python
import pandas as pd
```

```python
foo = pd.read_excel("hasil_3+tambahan.xlsx")
foo
```

```python
tanggal = foo['tanggal'].tolist()
barang  = foo['barang'].tolist()

temp_t = list()
temp_b = list()
temp_b_lower = list()
iterasi = 0
for i, j in zip(tanggal,barang):
    if i == "Laporan Data Penjualan Per Transaksi" or i == 'Maharani':
        i = tanggal[iterasi-1]
        if i == "Laporan Data Penjualan Per Transaksi" or i == 'Maharani':
            i = tanggal[iterasi-2]
    i = i.split("/")
    try:
        i[0],i[2] = i[2],i[0]
    except:
        print(iterasi)
        print(i)
    temp_t.append("-".join(i))
    j = j.replace("^","").replace("#","").replace("'","`").replace('"','^') #cleaning tahap 2
    temp_b.append(j)
    temp_b_lower.append(j.lower())
    iterasi+=1
dic = {
    "tanggal":temp_t,
    "barang":temp_b,
    "barang_lower":temp_b_lower
} 

df = pd.DataFrame(dic)
df.to_excel("hasil_akhir2.xlsx") #disimpan
```

```python
id = 1
file = open("insert_data.sql","w")
file.write("INSERT INTO `transaksi`(id, transaction_date,produk)\n")
file.write("VALUES\n")
for tgl, brg in zip(temp_t,temp_b):
    file.write('({},"{}","{}"),\n'.format(id, tgl, brg))
    id+=1
file.write(";")
file.close() 
```
